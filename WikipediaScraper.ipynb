```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
import requests
```

```python
print('What do you wanna search for today??')
title = str(input('Enter the title here: ')).replace(' ', '+')


link = 'https://www.google.co.in/search?q=' + title + '+wikipedia'
print(link)
res = requests.get(link)
print(res.text)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
print(soup)

for sp in soup.find_all('div'):
  try:
    link1 = sp.find('a').get('href')
    print(link1)
  except:
    pass
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Generative_artificial_intelligence')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.find('h1').text
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()
```

```python
print(corpus)
```

```python
for i in tqdm(range(1, 241)):
  textToReplace = f'[{i}]'
  corpus = corpus.replace(textToReplace, '')

print(corpus)
```
