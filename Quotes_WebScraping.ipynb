## Scrapping from Quotes to scrape

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
res = requests.get(link)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
data = []

for sp in soup.find_all('div', class_ = 'quote'):
  quote = sp.find('span', class_ = 'text')
  author = sp.find('small')
  author_id = sp.find('a').get('href')
  tags = []
  for tag in sp.find_all('a', class_ = 'tag'):
    tags.append(tag.text)
  tags = ', '.join(tags)
  data.append([quote.text[1: -1], author.text, author_id, tags])


df = pd.DataFrame(data, columns = ['quote', 'author', 'author_id', 'tags'])
df
```

```python
df.to_csv('quotes.csv')
```

### Scraping from multiple pages

```python
data = []

for page in tqdm(range(1, 11)):
  new_link = 'https://quotes.toscrape.com/page/' + str(page)
  res = requests.get(new_link)
  html = res.text
  soup = BeautifulSoup(html, 'html.parser')

  for sp in soup.find_all('div', class_ = 'quote'):
    quote = sp.find('span', class_ = 'text')
    author = sp.find('small')
    author_id = sp.find('a').get('href')
    tags = []
    for tag in sp.find_all('a', class_ = 'tag'):
      tags.append(tag.text)
    tags = ', '.join(tags)
    data.append([quote.text[1: -1], author.text, author_id, tags])


print(data)
df = pd.DataFrame(data, columns = ['quote', 'author', 'author_id', 'tags'])
df
```

```python
df['author_link'] = 'https://quotes.toscrape.com/author/' + df['author_id']
df



df.to_csv('quotes_multiple_pages.csv', index = False)
```
