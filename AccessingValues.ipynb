```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
df.head()
```

```python
impute = SimpleImputer(strategy='mean', missing_values=np.nan)
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
df.dropna()
```

##Analysing more than 1 column at a time

1. free apps in art and design

```python
df = df.values
```

```python
count = 0
for i in df:
  if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
    count += 1
print(count)
```

2. Art and design with rating more than 4.5

```python
count = 0
for i in df:
  if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5):
    count += 1
print(count)
```

2. Art and design with rating more than 4.5 and free

```python
count = 0
for i in df:
  if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5 and i[6] == 'Free'):
    count += 1
print(count)
```

2. Art and design with rating more than 4.5 and free but for which apps?

```python
for i in df:
  if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5 and i[6] == 'Free'):
    print(i[0], '\n')
```
